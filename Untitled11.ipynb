```python
"""
Electrical Device Identification and Failure Prediction System
Analyzing electrical waveforms for device recognition and fault prediction
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal as sp
from scipy.fft import fft, fftfreq
from scipy import stats
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

print("=" * 70)
print("ELECTRICAL DEVICE IDENTIFICATION & FAILURE PREDICTION SYSTEM")
print("=" * 70)

class SignalProcessor:
    """Advanced signal processing for electrical waveforms"""

    def __init__(self, sampling_rate=50000):
        self.sampling_rate = sampling_rate
        self.nyquist = sampling_rate / 2

    def process_pipeline(self, signal):
        """Complete signal processing pipeline"""
        # Detrend signal
        signal = sp.detrend(signal)

        # Apply bandpass filter (20Hz - 2000Hz)
        low = 20 / self.nyquist
        high = 2000 / self.nyquist
        b, a = sp.butter(4, [low, high], btype='band')
        signal = sp.filtfilt(b, a, signal)

        # Apply notch filter at 50Hz
        notch_freq = 50 / self.nyquist
        b, a = sp.iirnotch(notch_freq, Q=30)
        signal = sp.filtfilt(b, a, signal)

        # Normalize signal
        signal = (signal - np.median(signal)) / (np.percentile(signal, 75) - np.percentile(signal, 25))
        return np.clip(signal, -4, 4)

    def extract_features(self, signal):
        """Extract comprehensive features from signal"""
        features = {}

        # Time domain features
        features['RMS'] = float(np.sqrt(np.mean(signal**2)))
        features['Mean'] = float(np.mean(signal))
        features['Std'] = float(np.std(signal))
        features['Skewness'] = float(stats.skew(signal))
        features['Kurtosis'] = float(stats.kurtosis(signal))
        features['Peak'] = float(np.max(np.abs(signal)))
        features['Crest_Factor'] = float(features['Peak'] / features['RMS'] if features['RMS'] != 0 else 0)
        features['Energy'] = float(np.sum(signal**2))

        # Zero crossing rate
        features['Zero_Crossing_Rate'] = float(np.sum(np.diff(np.sign(signal)) != 0) / len(signal))

        # Frequency domain features
        if len(signal) > 50:
            n = len(signal)
            freq_domain = fft(signal)
            freqs = fftfreq(n, 1/self.sampling_rate)

            # Power ratios
            mask_low = (np.abs(freqs) >= 0) & (np.abs(freqs) < 100)
            mask_high = (np.abs(freqs) >= 1000)

            power_low = np.sum(np.abs(freq_domain[mask_low])**2)
            power_high = np.sum(np.abs(freq_domain[mask_high])**2)
            total_power = power_low + power_high

            if total_power > 0:
                features['Low_Freq_Ratio'] = float(power_low / total_power)
                features['High_Freq_Ratio'] = float(power_high / total_power)

        return features

class DeviceIdentifier:
    """Intelligent device identification system"""

    def __init__(self):
        self.processor = SignalProcessor()
        self.device_patterns = {
            'Refrigerator': {'rms': (1.5, 3.0), 'crest': (1.8, 2.5), 'low_freq': (0.6, 0.9)},
            'Air Conditioner': {'rms': (5.0, 8.0), 'crest': (2.5, 3.5), 'low_freq': (0.4, 0.7)},
            'Washing Machine': {'rms': (2.0, 4.0), 'crest': (2.0, 3.0), 'low_freq': (0.3, 0.6)},
            'LED Light': {'rms': (0.1, 0.5), 'crest': (1.2, 2.0), 'low_freq': (0.1, 0.3)},
            'Microwave Oven': {'rms': (4.0, 7.0), 'crest': (2.5, 4.0), 'low_freq': (0.2, 0.5)},
            'Industrial Motor': {'rms': (8.0, 15.0), 'crest': (3.0, 4.5), 'low_freq': (0.5, 0.8)}
        }

    def identify_device(self, signal):
        """Identify electrical device from waveform"""
        processed_signal = self.processor.process_pipeline(signal)
        features = self.processor.extract_features(processed_signal)

        scores = {}
        for device, pattern in self.device_patterns.items():
            score = 0
            total_weight = 0

            # Match RMS value
            if 'RMS' in features:
                rms = features['RMS']
                rmin, rmax = pattern['rms']
                if rmin <= rms <= rmax:
                    rms_score = 1.0 - abs(rms - (rmin + rmax)/2) / (rmax - rmin)
                    score += rms_score * 0.4
                total_weight += 0.4

            # Match Crest Factor
            if 'Crest_Factor' in features:
                crest = features['Crest_Factor']
                cmin, cmax = pattern['crest']
                if cmin <= crest <= cmax:
                    crest_score = 1.0 - abs(crest - (cmin + cmax)/2) / (cmax - cmin)
                    score += crest_score * 0.3
                total_weight += 0.3

            # Match frequency pattern
            if 'Low_Freq_Ratio' in features:
                ratio = features['Low_Freq_Ratio']
                lmin, lmax = pattern['low_freq']
                if lmin <= ratio <= lmax:
                    freq_score = 1.0 - abs(ratio - (lmin + lmax)/2) / (lmax - lmin)
                    score += freq_score * 0.3
                total_weight += 0.3

            if total_weight > 0:
                scores[device] = score / total_weight

        if scores:
            best_device = max(scores, key=scores.get)
            confidence = scores[best_device]

            # If confidence is low, classify by power level
            if confidence < 0.4:
                rms = features.get('RMS', 0)
                if rms < 1.0:
                    return "Small Electronic Device", 0.6, features
                elif rms < 3.0:
                    return "Medium Appliance", 0.5, features
                else:
                    return "High Power Device", 0.5, features

            return best_device, confidence, features

        return "Unknown Device", 0.0, features

class FailurePredictor:
    """Predict device failures from electrical signatures"""

    def __init__(self):
        self.health_thresholds = {
            'THD': {'warning': 0.1, 'critical': 0.2},
            'Crest_Factor': {'warning': 3.0, 'critical': 4.0},
            'High_Freq_Ratio': {'warning': 0.3, 'critical': 0.5},
            'Kurtosis': {'warning': 5.0, 'critical': 10.0}
        }

    def analyze_health(self, features):
        """Analyze device health from features"""
        issues = []
        warning_count = 0
        critical_count = 0

        # Check Total Harmonic Distortion (simulated)
        thd = abs(features.get('Skewness', 0)) * 0.05
        if thd > self.health_thresholds['THD']['critical']:
            issues.append(f"Critical: High harmonic distortion ({thd:.3f})")
            critical_count += 1
        elif thd > self.health_thresholds['THD']['warning']:
            issues.append(f"Warning: Moderate harmonic distortion ({thd:.3f})")
            warning_count += 1

        # Check Crest Factor
        crest = features.get('Crest_Factor', 0)
        if crest > self.health_thresholds['Crest_Factor']['critical']:
            issues.append(f"Critical: High crest factor ({crest:.2f})")
            critical_count += 1
        elif crest > self.health_thresholds['Crest_Factor']['warning']:
            issues.append(f"Warning: Elevated crest factor ({crest:.2f})")
            warning_count += 1

        # Check high frequency noise
        hf_ratio = features.get('High_Freq_Ratio', 0)
        if hf_ratio > self.health_thresholds['High_Freq_Ratio']['critical']:
            issues.append(f"Critical: Excessive high-frequency noise ({hf_ratio:.2f})")
            critical_count += 1
        elif hf_ratio > self.health_thresholds['High_Freq_Ratio']['warning']:
            issues.append(f"Warning: High-frequency noise present ({hf_ratio:.2f})")
            warning_count += 1

        # Determine overall status
        if critical_count > 0:
            status = "CRITICAL"
            fault_prob = min(0.95, 0.4 + critical_count * 0.2)
        elif warning_count > 1:
            status = "WARNING"
            fault_prob = min(0.7, 0.2 + warning_count * 0.15)
        elif warning_count > 0:
            status = "WARNING"
            fault_prob = 0.3
        else:
            status = "HEALTHY"
            fault_prob = 0.05

        return {
            'status': status,
            'fault_probability': fault_prob,
            'issues': issues,
            'critical_count': critical_count,
            'warning_count': warning_count
        }

class SystemVisualizer:
    """Create visualizations for analysis results"""

    @staticmethod
    def plot_waveform_comparison(signal, processed_signal, device_name):
        """Plot original vs processed waveform"""
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

        # Time domain
        t = np.linspace(0, 0.02, min(1000, len(signal)))
        ax1.plot(t, signal[:len(t)], 'b-', alpha=0.7, label='Original')
        ax1.plot(t, processed_signal[:len(t)], 'r-', alpha=0.7, label='Processed')
        ax1.set_title(f'{device_name} - Waveform Comparison')
        ax1.set_xlabel('Time (s)')
        ax1.set_ylabel('Amplitude')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # Frequency domain
        n = len(signal)
        freqs = fftfreq(n, 1/50000)[:n//2]
        orig_fft = np.abs(fft(signal)[:n//2])
        proc_fft = np.abs(fft(processed_signal)[:n//2])

        ax2.plot(freqs[:500], orig_fft[:500], 'b-', alpha=0.5, label='Original')
        ax2.plot(freqs[:500], proc_fft[:500], 'r-', alpha=0.7, label='Processed')
        ax2.set_title('Frequency Spectrum')
        ax2.set_xlabel('Frequency (Hz)')
        ax2.set_ylabel('Magnitude')
        ax2.legend()
        ax2.grid(True, alpha=0.3)

        plt.tight_layout()
        return fig

    @staticmethod
    def plot_analysis_report(device, confidence, health_result, features):
        """Create comprehensive analysis report"""
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(14, 10))

        # Device identification
        colors = {'HEALTHY': 'green', 'WARNING': 'orange', 'CRITICAL': 'red'}
        status_color = colors.get(health_result['status'], 'gray')

        ax1.text(0.5, 0.7, f'Device: {device}', fontsize=16, ha='center')
        ax1.text(0.5, 0.5, f'Confidence: {confidence:.1%}', fontsize=14, ha='center')
        ax1.text(0.5, 0.3, f'Status: {health_result["status"]}',
                fontsize=18, ha='center', color=status_color, fontweight='bold')
        ax1.axis('off')

        # Fault probability gauge
        fault_percent = health_result['fault_probability'] * 100
        ax2.barh(0, fault_percent, color='red', height=0.3)
        ax2.set_xlim(0, 100)
        ax2.set_title(f'Fault Probability: {fault_percent:.1f}%')
        ax2.set_xlabel('Probability (%)')
        ax2.grid(True, alpha=0.3)

        # Key features bar chart
        key_features = ['RMS', 'Crest_Factor', 'Kurtosis', 'Zero_Crossing_Rate']
        values = [features.get(f, 0) for f in key_features]
        labels = ['RMS', 'Crest\nFactor', 'Kurtosis', 'Zero\nCrossing']

        bars = ax3.bar(range(len(values)), values)
        ax3.set_xticks(range(len(values)))
        ax3.set_xticklabels(labels)
        ax3.set_title('Key Features')
        ax3.grid(True, alpha=0.3, axis='y')

        # Health issues
        if health_result['issues']:
            issues_text = '\n'.join(health_result['issues'])
            ax4.text(0.1, 0.5, issues_text, fontsize=11,
                    verticalalignment='center', wrap=True)
        else:
            ax4.text(0.5, 0.5, 'No issues detected',
                    fontsize=14, ha='center', va='center')
        ax4.set_title('Health Indicators')
        ax4.axis('off')

        plt.suptitle(f'Device Analysis Report - {datetime.now().strftime("%Y-%m-%d %H:%M")}',
                    fontsize=16, fontweight='bold')
        plt.tight_layout()
        return fig

class ElectricalAnalyzer:
    """Main system controller"""

    def __init__(self):
        self.identifier = DeviceIdentifier()
        self.predictor = FailurePredictor()
        self.visualizer = SystemVisualizer()
        self.processor = SignalProcessor()

    def generate_test_signal(self, device_type, noise_level=0.1):
        """Generate test signal for a device type"""
        duration = 0.1  # 100ms
        t = np.linspace(0, duration, int(50000 * duration))

        # Base signals for different devices
        if device_type == "Refrigerator":
            signal = 2.0 * np.sin(2 * np.pi * 50 * t)
            signal += 0.3 * np.sin(2 * np.pi * 150 * t)
        elif device_type == "Air Conditioner":
            signal = 6.0 * np.sin(2 * np.pi * 50 * t)
            signal += 1.0 * np.sin(2 * np.pi * 100 * t)
        elif device_type == "LED Light":
            signal = 0.3 * np.sin(2 * np.pi * 1000 * t)
            signal += 0.1 * np.sin(2 * np.pi * 10000 * t)
        elif device_type == "Industrial Motor":
            signal = 10.0 * np.sin(2 * np.pi * 50 * t)
            signal += 2.0 * np.sin(2 * np.pi * 150 * t)
        else:
            signal = 1.0 * np.sin(2 * np.pi * 50 * t)

        # Add noise
        signal += noise_level * np.random.randn(len(t))

        return signal

    def analyze_device(self, signal=None, device_type="Refrigerator"):
        """Complete analysis of an electrical device"""
        if signal is None:
            signal = self.generate_test_signal(device_type)

        print(f"\n{'='*50}")
        print(f"ANALYZING DEVICE SIGNAL")
        print(f"{'='*50}")

        # Process signal
        processed = self.processor.process_pipeline(signal)

        # Identify device
        device, confidence, features = self.identifier.identify_device(signal)
        print(f"Identified: {device}")
        print(f"Confidence: {confidence:.1%}")

        # Predict failures
        health = self.predictor.analyze_health(features)
        print(f"Health Status: {health['status']}")
        print(f"Fault Probability: {health['fault_probability']:.1%}")

        # Display issues
        if health['issues']:
            print("\nHealth Issues Found:")
            for issue in health['issues']:
                print(f"  • {issue}")

        # Generate visualizations
        fig1 = self.visualizer.plot_waveform_comparison(signal, processed, device)
        fig2 = self.visualizer.plot_analysis_report(device, confidence, health, features)

        # Save results
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        fig1.savefig(f'waveform_{timestamp}.png', dpi=150, bbox_inches='tight')
        fig2.savefig(f'analysis_{timestamp}.png', dpi=150, bbox_inches='tight')

        print(f"\nResults saved with timestamp: {timestamp}")

        return {
            'device': device,
            'confidence': confidence,
            'health_status': health['status'],
            'fault_probability': health['fault_probability'],
            'features': features,
            'issues': health['issues']
        }

# Main execution
if __name__ == "__main__":
    print("\n" + "="*70)
    print("ELECTRICAL DEVICE IDENTIFICATION SYSTEM")
    print("="*70)

    # Initialize system
    analyzer = ElectricalAnalyzer()

    # Test with different devices
    test_devices = ["Refrigerator", "Air Conditioner", "LED Light", "Industrial Motor"]

    all_results = []

    for device in test_devices:
        print(f"\n▶ Testing: {device}")
        result = analyzer.analyze_device(device_type=device)
        all_results.append(result)

    # Summary report
    print(f"\n{'='*70}")
    print("SUMMARY REPORT")
    print(f"{'='*70}")

    for i, result in enumerate(all_results):
        print(f"\n{i+1}. {result['device']}")
        print(f"   Status: {result['health_status']}")
        print(f"   Fault Risk: {result['fault_probability']:.1%}")

    # Show one plot
    plt.show()


    print(f"\n✅ Analysis complete! Check generated PNG files for results.")
```
